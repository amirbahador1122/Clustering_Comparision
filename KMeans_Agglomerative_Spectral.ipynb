```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering
```

```python
data = pd.read_csv('spiral.txt', header = None, delimiter = "\t")
```

```python
data.head()
```

```python
X = data[[0,1]]
```

```python
plt.scatter(X[0], X[1])
```

```python
kmeans_pred = KMeans(n_clusters=3).fit_predict(X)
```

```python
plt.scatter(X[0], X[1], c=kmeans_pred)
```

```python
agglomerative_pred_complete = AgglomerativeClustering(n_clusters=3, linkage='complete').fit_predict(X)
```

```python
plt.scatter(X[0], X[1], c=agglomerative_pred_complete)
```

```python
agglomerative_pred_avg = AgglomerativeClustering(n_clusters=3, linkage='average').fit_predict(X)
```

```python
plt.scatter(X[0], X[1], c=agglomerative_pred_avg)
```

```python
agglomerative_pred_single = AgglomerativeClustering(n_clusters=3, linkage='single').fit_predict(X)
```

```python
plt.scatter(X[0], X[1], c=agglomerative_pred_single)
```

```python
spectral_pred = SpectralClustering(n_clusters=3, affinity = 'rbf').fit_predict(X)
```

```python
plt.scatter(X[0], X[1], c=spectral_pred)
```

```python
spectral_pred = SpectralClustering(n_clusters=3, affinity = 'rbf').fit_predict(X)
```

```python
plt.scatter(X[0], X[1], c=spectral_pred)
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)

ax1.scatter(X[0], X[1])
ax1.set_title('Before Clustering')
ax2.scatter(X[0], X[1], c=spectral_pred)
ax2.set_title('Spectral Clustering with RBF Kernel')
ax3.scatter(X[0], X[1], c=kmeans_pred)
ax3.set_title('KMeans Clustering')
ax4.scatter(X[0], X[1], c=agglomerative_pred_avg)
ax4.set_title('Agglomerative Clustering - Average linkage')
ax5.scatter(X[0], X[1], c=agglomerative_pred_complete)
ax5.set_title('Agglomerative Clustering - Complete linkage')
ax6.scatter(X[0], X[1], c=agglomerative_pred_single)
ax6.set_title('Agglomerative Clustering - Single linkage')
plt.show()
```
